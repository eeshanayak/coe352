```python
for i in dt_list:
    dt = i
    num_entries = int(k/dt+1)
    temp_u = np.empty(num_entries) # positions
    temp_v = np.empty(num_entries) # velocities

    # initial conditions 
    t = 0
    temp_u[0] = 1
    temp_v[0] = 0
    anew = f(temp_u[0], t)
    
    for j in range(0, int(k/dt)):
        t = t + dt
        aold = anew
        temp_u[j] = temp_u[j-1] + temp_v[j-1]*h + 0.5*aold*h**2
        anew = f(temp_u[j], t)
        temp_v[j] = temp_v[j-1] + 0.5*(aold + anew)*h
    
    temp_tspan = np.arange(0, k + dt, dt)
    
    tspan.append(temp_tspan)
    u.append(temp_u)
    v.append(temp_v)
```

```python
for i in range (len(dt_list)):
    line_label = "dt = " + str(dt_list[i])
    plt.plot(tspan[i], u[i], label = line_label)
    plt.legend()
    
plt.show()
```
